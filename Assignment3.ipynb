```python
import pandas as pd
df = pd.read_csv("nba.csv")
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
mode = df['College'].mode()[0]

df['College'].fillna(mode, inplace = True)
```

```python
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace = True)
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
df['Height'].value_counts()
```

```python
group_by_height = df.groupby(df['Height'])
group_by_height.get_group('6-11')
```

```python
group_by_height['Salary'].describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
c = group_by_height['Salary'].describe()['count']
plt.bar(c.index, c.values)
plt.show()

plt.figure(figsize=(10,6))
m = group_by_height['Salary'].describe()['mean']
plt.bar(m.index, m.values)
plt.show()

plt.figure(figsize=(10,6))
s = group_by_height['Salary'].describe()['std']
plt.bar(s.index, s.values)
plt.show()
```

```python
df['Age'].value_counts().hist()
plt.show()
```

```python
df['age_group'] = pd.cut(df['Age'], bins = [19,24, 29, 34, 40], labels = ['19-24', '24-29', '29-34', '34-40'])
df.head()
```

```python
df['age_group'].value_counts()
```

```python
age_grouped_df = df.groupby(df['age_group'])
age_grouped_df.get_group('34-40')
```

```python
age_grouped_df['Salary'].describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
c = age_grouped_df['Salary'].describe()['count']
plt.bar(c.index, c.values)
plt.show()

plt.figure(figsize=(10,6))
m = age_grouped_df['Salary'].describe()['mean']
plt.bar(m.index, m.values)
plt.show()

plt.figure(figsize=(10,6))
s = age_grouped_df['Salary'].describe()['std']
plt.bar(s.index, s.values)
plt.show()
```

```python
list_of_salaries = list(df.groupby('age_group')['Salary'])
list_of_salaries
```
